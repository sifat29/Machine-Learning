```python
tp = float(input("Enter your True postive value: "))
tn = float(input("Enter your True negetive value: "))
fp = float(input("Enter your False postive value: "))
fn = float(input("Enter your False negetive value: "))
```

```python
import numpy as np
x=np.array([[tp,fn], [fp,tn]])
print(x)
```

```python
Accuracy =((tp+tn)/(tp+fp+fn+tn)* 100)
print("The Accuracy is", Accuracy)
```

```python
Precision = (tp/(tp+fp)*100)
print("The precision is", Precision)
```

```python
Recall = (tp/(tp+fn)*100)
print("The precision is", Recall)
```

```python
F1_Score = (2*(Recall * Precision) / (Recall + Precision)*100)
print("The precision is", F1_Score)
```

```python
Sensitivity = (tp / (tp + fn)*100)
print("The precision is", Sensitivity)
```

```python
Specificity = (tn / (fp + tn)*100)
print("The precision is", Specificity)
```
